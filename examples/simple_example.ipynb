```python
import matplotlib.pyplot as plt
import pynrn as nrn
```

```python

# Create a simulation context that will contain all items to simulate
sim = nrn.Context()

# create a single compartment neuron
cell = nrn.Section()
cell.insert('hh')
cell.insert('pas')

# Create an IClamp to stimulate the cell
stim = nrn.IClamp(cell(0.5), delay=5, dur=10, amp=45)

# Record the time and membrane potential
t_rec = nrn.Vector(sim.t)
v_rec = nrn.Vector(cell(0.5).v)

# Run the simulation
sim.run(dt=0.025, tstop=30)

plt.plot(t_rec, v_rec)
plt.xlabel('Time (ms)')
plt.ylabel('Membrane potential (mV)');
```
