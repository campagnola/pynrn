```python
import faulthandler
faulthandler.enable()
import pynrn.allen_model
import pynrn
from neuron import h
```

```python
with pynrn.Context() as sim:
    duration = 100
    dt = 0.025
    sections = pynrn.allen_model.load_allen_cell(489932682)  # Mouse VisP L5 pyramidal (perisomatic model)
    soma = sections['soma[0]']
    ic = pynrn.IClamp(soma(0.5), delay=20, dur=60, amp=.5)
    vm = pynrn.Vector(soma(0.5).v)
    sim.run(dt=dt, celsius=25, tstop=duration)

```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(len(vm)) * dt
plt.plot(t, vm)
```
